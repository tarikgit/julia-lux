```julia
# Avant 2006, Unité: mm. À partir de 2006, Unité: millibar.
#
# La moyenne séculaire de la pression atmosphérique observée à Luxembourg-Ville réduite à 0 degré centigrade, est de 733.5 mm. En considérant que les lectures ont été faites à la température moyenne, la valeur précitée, ramenée au niveau de la mer, s'élève à 733.5 + 29.0 = 762.5 mm ou 1 016.6 millibar.
#
# Selon les moyennes disponibles la moyenne journalière est calculée:
# ­ par un appareil enregistreur
# ­ en divisant par 24 la somme des observations faites heure par heure
# ­ en divisant par 3 la somme des observations faites à 07.00, 13.00 et 21.00 heures.
# La moyenne du mois est obtenue par addition des moyennes de tous les jours et division par le nombre de jours du mois etc. (voir LAHR, Bulletin du STATEC 1966, No 7).
#
p5 = Plots.plot( dr_pressure, data_pressure[3] , title="Pression atmosphérique à Luxembourg-Ville", label=["pression athmosph."] )
```

```julia
### E2607 Heures travaillées total (HRT) par branche (NaceR2) (en 1 000 d'heures) 1995 - 2018
### E2500 Principaux agrégats: trois approches (prix courants) (en millions EUR) 1995 - 2018
### STATEC
using CSV
data_productivity_raw = CSV.read("/Users/tarikz/Documents/GitHub/julia-lux/data/e2607-e2500.csv", datarow = 2, delim = ',')
```

```julia
### Productivity crisis: Should Luxembourg be concerned?
### Method 1
# Define productivity: (Produit Intérieur Brut au prix du marché) / (Heures travaillées total en 1 000 d'heures)
productivity_lux = data_productivity_raw.e2500./data_productivity_raw.e2607
# Construct dates (quarters)
using Dates
dr_productivity_lux = Dates.Date(1995,1):Dates.Month(3):Dates.Date(2018,4)
# Plot productivity fro Luxembourg
using Plots
Plots.plot( dr_productivity_lux, productivity_lux, title = " Labour productivity", label = "" )
### Notes:
# Productivity trend stagnates but becomes more volatile. Are there similarities to the 'UK productivity trend'?

# I used GDP in nominal terms, I have to try it in real terms.
```

```julia
### Method 2
### STATEC
using CSV
data_ipcn = CSV.read("/Users/tarikz/Documents/GitHub/julia-lux/data/e5101.csv", datarow = 2, delim = ',')
# Construct dates (quarters)
using Dates
dr_ipcn = Dates.Date(1995,1):Dates.Month(1):Dates.Date(2018,10)
```

```julia
# Plot ipcn for Luxembourg 
# E5101 Moyennes semestrielles des indices (IPCN Base au 1.1.1948) 1948 - 2018
using Plots
Plots.plot( dr_ipcn, data_ipcn.ipcn, title = " Moyennes semestrielles des indices (IPCN Base au 1.1.1948) 1948 - 2018", label = "" )
```

```julia
# Convert monthly data into quarterly

nrows_monthly, ncols_monthly = size(data_ipcn)

nrows_monthly, ncols_monthly = size(data_ipcn)
nrows_qrtly = Int( floor( nrows_monthly/3 ) )

a_qrtly = ones( nrows_qrtly )

for i in 1:nrows_qrtly
    a_qrtly[i] = data_ipcn.ipcn[i*3,1]
end

# Growth rate of prices
b_qrtly = ones( nrows_qrtly )

for j in 2:nrows_qrtly
    b_qrtly[j] = ( a_qrtly[j]-a_qrtly[j-1] ) / a_qrtly[j-1]    
end
b_qrtly

# Deflate the productivity
productivity_lux_defl = productivity_lux./(ones(size(b_qrtly))-b_qrtly)[2:95]

# Plot productivity fro Luxembourg
using Plots
Plots.plot( dr_productivity_lux, [productivity_lux_defl productivity_lux], title = " Labour productivity (deflated)", label = "" )
```
