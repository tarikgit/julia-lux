```julia
using JuMP
m = Model()
@variable(m, x, start = 0.0)
@variable(m, y, start = 0.0)

@NLobjective(m, Min, (1-x)^2 + 100(y-x^2)^2)

solve(m)
println("x = ", getvalue(x), " y = ", getvalue(y))

# adding a (linear) constraint
@constraint(m, x + y == 10)
solve(m)
println("x = ", getvalue(x), " y = ", getvalue(y))
```

```julia
using JuMP

using Ipopt


"""
The problem that I want to solve has 4 variables and 6 constraints.
It is the following:


min_x = x1x4(x1+x2+x3) + x3


s.t.

x1*x2*x3*x4 >= 25
x_1^2 + x_2^2 + x_3^2 + x_4^2 = 40
1 <= x1,x2,x3,x4 <= 5


starting values: x0 = (x1 = 1, x2 = 5, x3 = 5, x4 = 1)
"""


#tic()


m = Model(solver=IpoptSolver())


@variable(m, 1 <= x1 <= 5)
@variable(m, 1 <= x2 <= 5)
@variable(m, 1 <= x3 <= 5)
@variable(m, 1 <= x4 <= 5)

@NLobjective(m, Min,  x1 * x4 * (x1 + x2 + x3) + x3)
@NLconstraint(m, x1^2 + x2^2 + x3^2 + x4^2 == 40)
@NLconstraint(m, x1 * x2 * x3 * x4 >= 25)


setvalue(x1, 1)
setvalue(x2, 5)
setvalue(x3, 5)
setvalue(x4, 1)

status = solve(m)

println("got ", getobjectivevalue(m), " at ", [getvalue(x1),getvalue(x2), 
getvalue(x3), getvalue(x4)])

#toc()
```

```julia

using JuMP

using Ipopt


distance = [20, 30, 50, 100]
speed = [5, 10, 20, 30]

x = distance
y = speed


m = Model(solver=IpoptSolver())


@variable(m, 0 <= a <= 5)
@variable(m, 0 <= b <= 5)

@NLobjective(m, Min,  [y-a.*log.( b.*x )]'*[y-a.*log.( b.*x )] )
@NLconstraint(m, b.*x != 0)
#@NLconstraint(m, x1 * x2 * x3 * x4 >= 25)

# Initialisation of the parameteres
setvalue(a, 1)
setvalue(b, 1)

status = solve(m)

println("got ", getobjectivevalue(m), " at ", [getvalue(a),getvalue(b)])
```
